```python
import seaborn as sns
import matplotlib.pyplot as plt

# Set theme and figure size
sns.set_theme(style="ticks")
plt.figure(figsize=(10, 6))

# Draw Scatter Plot
sns.scatterplot(
    data=data,
    x="มูลค่างานหลัง Vat",
    y="มูลค่างาน (รวม VAT) - มูลค่าสุทธิ",
    hue="สถานะ",  # Color points by document status
    s=100,
    alpha=0.8,
    palette="Set2"
)

# Setup Title and Labels in English
plt.title("Comparison of Work Value (After VAT) vs Net Work Value (incl. VAT)", fontsize=14, fontweight="bold")
plt.xlabel("Work Value After VAT", fontsize=12)
plt.ylabel("Net Work Value (incl. VAT)", fontsize=12)
plt.grid(True, linestyle="--", alpha=0.5)

# Setup Legend
plt.legend(title="Document Status")

plt.tight_layout()
plt.show()
```
